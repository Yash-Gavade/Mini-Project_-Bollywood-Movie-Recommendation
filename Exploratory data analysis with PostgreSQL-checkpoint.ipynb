```python
#sql statement to create table
sqlstatement='''CREATE TABLE Movies(
              userId INT,
              Movie_id INT,
              Movie_Title text,
              Released_Year INT,
              Duration VARCHAR,
              Genre text,
              PRIMARY KEY(Movie_id)
              )'''
              
db.create_table(sqlstatement, ' Movies')
#sql statement to create table
sqlstatement='''CREATE TABLE Rating(
              userId INT,
              Movie_id INT,
              Movie_Title VARCHAR,
              Rating VARCHAR,	
              votes VARCHAR,

              PRIMARY KEY(Movie_id,userId),
              FOREIGN KEY(Movie_id) 
                  REFERENCES Movies(Movie_id)
              )'''
db.create_table(sqlstatement, ' Ratings')

```

## Loading Data into tables from dataFrames
After creating my tables now I've to load data from data frames into tables


```python

cursor = db.cursor
for row in moviesData.itertuples():
    cursor.execute('''
                        INSERT INTO Movies(userId,Movie_id,Movie_Title,Released_Year,Duration,Genre)
                        VALUES (?,?,?,?,?,?)
                        ''',
                        row.userId,
                        row.Movie_id,
                        row.Movie_Title,
                        row.Released_Year,
                        row.Duration,
                        row.Genre
  
                  )
db.db_connection.commit()
print("successfully inserted data from DataFrame into table")

for row in ratingsData.itertuples():
    cursor.execute('''
                        INSERT INTO Rating(userId,Movie_id,Movie_Title,Rating,votes)
                        VALUES (?,?,?,?,?)
                        ''',
                        row.userId,
                        row.Movie_id,
                        row.Movie_Title,
                        row.Rating,
                        row.votes                        

                  )
db.db_connection.commit()
print("successfully inserted data from DataFrame into table")
```

```python
#taking Join of tables(Rating and movies) then creating View(MoviesWithRatings) of result
command ='''
CREATE VIEW MoviesWithRatings AS
    SELECT Rating.userId,Rating.Movie_id,Rating.Movie_Title,Rating.rating,votes
    FROM Rating
    LEFT JOIN Movies
    ON Rating.Movie_id = Movies.Movie_id;
'''
db.run_command(command)
```

```python
#executing query over view 
query = '''
SELECT Movie_Title,votes AS movies_count
FROM MoviesWithRatings
GROUP BY Movie_Title,votes
ORDER BY movies_count DESC
LIMIT 20;
'''
query_result=db.run_query(query)
query_result
```

```python
plt.figure(figsize=(10, 5))
barh = plt.barh(query_result['movie_title'], query_result['movies_count'], color='blue')
plt.title(' top ten most rated movies ')
plt.xticks(np.arange(0, 400, 50))
plt.tight_layout()
```

```python
genres_unique = pd.DataFrame(moviesData.Genre.str.split('|').tolist()).stack().unique()
genres_list=genres_unique[:len(genres_unique)-1]
genres_count=[]
```

```python
for genre in genres_list:
    query = '''
    SELECT count(Movies.Movie_id) 
    FROM  Movies
    where Movies.Genre LIKE '%'''+genre+'''%'
    '''
    query_result=db.run_query(query)
    genres_count.append(query_result.iloc[0]['count'])
```

```python
data=pd.DataFrame({'x': genres_list, 'y': genres_count})
```

```python
data=pd.DataFrame({'x': genres_list, 'y': genres_count})
plt.subplots(figsize=(50,8))
sns.barplot(x=data.x, y=data.y, palette="rocket")
plt.xticks(rotation=-45)
plt.title('Most popular genres among directors');
```
