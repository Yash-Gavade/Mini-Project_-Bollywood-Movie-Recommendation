# EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
a = plt.cm.cool

plt.figure(figsize=(15,10))
count = imdb_movies['Director'].value_counts()[:20]
sns.barplot(count.values, count.index, palette=[a(0.1),a(0.2),a(0.3),a(0.4),a(0.5),a(0.6),a(0.7),a(0.8),a(0.9),a(0.99)])
for i, v in enumerate(count.values):
    plt.text(0.8,i,v,color='k',fontsize=14)
plt.xlabel('Count')
plt.ylabel('Production Company name')
plt.title("Distribution of Studio names")
```

```python
a = plt.cm.autumn

plt.figure(figsize=(15,10))
count = imdb_movies['Genre'].value_counts()[:10]
sns.barplot(count.values, count.index, palette=[a(0.1),a(0.2),a(0.3),a(0.4),a(0.5),a(0.6),a(0.7),a(0.8),a(0.9),a(0.99)])
for i, v in enumerate(count.values):
    plt.text(0.8,i,v,color='k',fontsize=14)
plt.xlabel('Count')
plt.ylabel('Genre')
plt.title("Distribution of Genres")
```

```python
num_data = ['Rating',  'Votes']
imdb_movies[num_data].describe()
```

```python

# Grouping the data based on Decades


movies_by_decades = imdb_movies[['Movie_id', 'Movie_Title', 'Released_Year','Rating', 'Votes']]
decades = movies_by_decades['Released_Year']//10*10
Decades = (decades+9).astype(str)
decades_column = pd.DataFrame(decades)

```

```python
movies_by_decades.insert(3, 'Decades', decades_column)
movies_by_decades.head()
```

```python
max_width = 15
fig, ax = plt.subplots(figsize = (16,5))
decades = movies_by_decades.groupby('Decades')['Movie_id'].count().index
count = movies_by_decades.groupby('Decades')['Movie_id'].count()
sns.barplot(ax = ax, x = decades, y = count)
ax.set_title('Movie Counts Based on Decade', fontsize = 18)
ax.set_xlabel('Decade')
for index,count in enumerate(count.astype(int)):
       ax.text(x=index+0.15 , y =count+1 , s=f"{count}" , fontdict=dict(fontsize=10))
ax.set_ylabel('No. of Movies')
plt.show()
```

## Genres and Avg_Votes

```python
movies_Genre = imdb_movies[['Movie_id', 'Movie_Title', 'Genre', 'Votes']]
movies_Genre['Genre'] = movies_Genre['Genre'].astype('str')

Genre_split = pd.DataFrame(movies_Genre['Genre'].str.split(',').tolist(), index=movies_Genre['Movie_id']).stack()
Genre_split = Genre_split.reset_index(['Movie_id'])
Genre_split.columns = ['Movie_id', 'Genre_split']
movies_Genre_split = pd.merge(Genre_split, movies_Genre[['Movie_id', 'Movie_Title', 'Votes']],
                              left_on = 'Movie_id', right_on = 'Movie_id')
movies_Genre_split['Genre_split'] = movies_Genre_split['Genre_split'].str.lstrip(' ').str.rstrip(' ')
movies_Genre_split.tail()
```
